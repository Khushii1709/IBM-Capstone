### Task 1: Trends in Compensation Over Age Groups


##### 1. Line Chart of Median `ConvertedCompYearly` by Age Group


- Track how the median yearly compensation (ConvertedCompYearly) changes across different age groups.

- Use a line chart to visualize these trends.


```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Connect to the SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Step 2: Query Age and ConvertedCompYearly
QUERY = """
SELECT Age, ConvertedCompYearly
FROM main
WHERE Age IS NOT NULL AND ConvertedCompYearly IS NOT NULL
"""
df = pd.read_sql_query(QUERY, conn)

# Step 3: Close the connection
conn.close()

# Step 4: Remove unrealistic outliers
df = df[df['ConvertedCompYearly'] <= 500000]

# Step 5: Define an ordered list for age groups
age_order = [
    'Under 18 years old', '18-24 years old', '25-34 years old', '35-44 years old',
    '45-54 years old', '55-64 years old', '65 years or older', 'Prefer not to say'
]

# Step 6: Calculate median compensation per age group
median_comp = df.groupby('Age')['ConvertedCompYearly'].median().reindex(age_order)

# Step 7: Plot the line chart
plt.figure(figsize=(12,6))
plt.plot(median_comp.index, median_comp.values, marker='o', linestyle='-', color='blue')
plt.title("Median Yearly Compensation by Age Group", fontsize=14)
plt.xlabel("Age Group")
plt.ylabel("Median Compensation (USD)")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

##### 2. Line Chart of Median `ConvertedCompYearly` for Ages 25 to 45


For a closer look, plot a line chart focusing on the median compensation for respondents between ages 25 and 45.


```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Connect to the SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Step 2: Query Age and ConvertedCompYearly
QUERY = """
SELECT Age, ConvertedCompYearly
FROM main
WHERE Age IS NOT NULL AND ConvertedCompYearly IS NOT NULL
"""
df = pd.read_sql_query(QUERY, conn)

# Step 3: Close the connection
conn.close()

# Step 4: Filter unrealistic compensation values
df = df[df['ConvertedCompYearly'] <= 500000]

# Step 5: Focus on age groups 25-34, 35-44
age_subset = ['25-34 years old', '35-44 years old']
df_filtered = df[df['Age'].isin(age_subset)]

# Step 6: Calculate median compensation per age group
median_comp = df_filtered.groupby('Age')['ConvertedCompYearly'].median().reindex(age_subset)

# Step 7: Plot the line chart
plt.figure(figsize=(8,5))
plt.plot(median_comp.index, median_comp.values, marker='o', linestyle='-', color='green')
plt.title("Median Yearly Compensation (Ages 25-45)", fontsize=14)
plt.xlabel("Age Group")
plt.ylabel("Median Compensation (USD)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Task 2: Trends in Job Satisfaction by Experience Level



##### 1. Line Chart of Job Satisfaction (`JobSatPoints_6`) by Experience Level



- Use a column that approximates experience level to analyze how job satisfaction changes with experience.

- If needed, substitute an available experience-related column for `Experience`.


```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Connect to the SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Step 2: Query YearsCodePro and JobSatPoints_6
QUERY = """
SELECT YearsCodePro, JobSatPoints_6
FROM main
WHERE YearsCodePro IS NOT NULL AND JobSatPoints_6 IS NOT NULL
"""
df = pd.read_sql_query(QUERY, conn)

# Step 3: Close the connection
conn.close()

# Step 4: Clean and convert YearsCodePro to numeric
# Some entries may be like "Less than 1 year" or "More than 50 years"
def convert_years(x):
    if x == "Less than 1 year":
        return 0.5
    elif x == "More than 50 years":
        return 51
    try:
        return float(x)
    except:
        return None

df['YearsCodeProNumeric'] = df['YearsCodePro'].apply(convert_years)
df = df.dropna(subset=['YearsCodeProNumeric', 'JobSatPoints_6'])

# Step 5: Group by experience level (YearsCodeProNumeric) and calculate mean job satisfaction
experience_groups = df.groupby('YearsCodeProNumeric')['JobSatPoints_6'].mean()

# Step 6: Plot the line chart
plt.figure(figsize=(10,5))
plt.plot(experience_groups.index, experience_groups.values, marker='o', linestyle='-', color='blue')
plt.title("Average Job Satisfaction vs. Years of Professional Coding Experience", fontsize=14)
plt.xlabel("Years of Professional Coding Experience")
plt.ylabel("Average Job Satisfaction (JobSatPoints_6)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Task 3: Trends in Job Satisfaction and Compensation by Experience


##### 1.Line Chart of Median ConvertedCompYearly Over Experience Level

- This line chart will track how median compensation (`ConvertedCompYearly`) changes with increasing experience.

- Use a column such as `WorkExp` or another relevant experience-related column.


```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Connect to SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Step 2: Query YearsCodePro and ConvertedCompYearly
QUERY = """
SELECT YearsCodePro, ConvertedCompYearly
FROM main
WHERE YearsCodePro IS NOT NULL AND ConvertedCompYearly IS NOT NULL
"""
df = pd.read_sql_query(QUERY, conn)

# Step 3: Close the database connection
conn.close()

# Step 4: Convert YearsCodePro to numeric values
def convert_years(x):
    if x == "Less than 1 year":
        return 0.5
    elif x == "More than 50 years":
        return 51
    try:
        return float(x)
    except:
        return None

df['YearsCodeProNumeric'] = df['YearsCodePro'].apply(convert_years)

# Step 5: Remove any rows with missing or unrealistic compensation values
df = df.dropna(subset=['YearsCodeProNumeric', 'ConvertedCompYearly'])
df = df[df['ConvertedCompYearly'] <= 500000]  # Optional: remove outliers

# Step 6: Calculate median compensation for each experience level
median_comp = df.groupby('YearsCodeProNumeric')['ConvertedCompYearly'].median().reset_index()

# Step 7: Plot the line chart
plt.figure(figsize=(12,6))
plt.plot(median_comp['YearsCodeProNumeric'], median_comp['ConvertedCompYearly'], marker='o', color='blue')
plt.title("Median Converted Compensation vs. Years of Professional Coding Experience")
plt.xlabel("Years of Professional Coding Experience")
plt.ylabel("Median Yearly Compensation (USD)")
plt.grid(True)
plt.show()
```
