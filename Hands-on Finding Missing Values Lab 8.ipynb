<h2>1. Load the Dataset</h2>

The functions below will download the dataset into your browser:



```python
# Define the URL of the dataset
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"

# Load the dataset into a DataFrame
df = pd.read_csv(file_path)

# Display the first few rows to ensure it loaded correctly
print(df.head())
```

### 2. Explore the Dataset
##### Task 1: Display basic information and summary statistics of the dataset.


```python
# 1. Display basic information about the dataset
print("Dataset Information:")
print(df.info())

# 2. Display summary statistics for numeric columns
print("\nSummary Statistics (Numeric Columns):")
print(df.describe())

# 3. Display summary statistics for non-numeric columns
print("\nSummary Statistics (Non-Numeric Columns):")
print(df.describe(include=['object']))
```

### 3. Finding Missing Values
##### Task 2: Identify missing values for all columns.


```python
# 1. Count missing values per column
missing_values = df.isnull().sum()

# 2. Display missing values for all columns
print("Missing values per column:")
print(missing_values)

# 3. Optionally, display only columns with missing values
print("\nColumns with missing values:")
print(missing_values[missing_values > 0])
```

##### Task 3: Visualize missing values using a heatmap (Using seaborn library).



```python
# Import seaborn and matplotlib
import seaborn as sns
import matplotlib.pyplot as plt

# Set the figure size
plt.figure(figsize=(15,8))

# Create a heatmap to visualize missing values
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')

# Add a title
plt.title("Heatmap of Missing Values in the Dataset")

# Show the plot
plt.show()
```

##### Task 4: Count the number of missing rows for a specific column (e.g., `Employment`).


```python
# Count missing values in the 'Employment' column
missing_employment = df['Employment'].isnull().sum()

# Display the result
print(f"Number of missing rows in 'Employment' column: {missing_employment}")
```

### 4. Imputing Missing Values
##### Task 5: Identify the most frequent (majority) value in a specific column (e.g., `Employment`).


```python
# Find the most frequent value (mode) in the 'Employment' column
most_frequent_value = df['Employment'].mode()[0]

# Display the result
print(f"The most frequent value in 'Employment' column is: {most_frequent_value}")
```

##### Task 6: Impute missing values in the `Employment` column with the most frequent value.



```python
# Impute missing values without using inplace to avoid FutureWarning
most_frequent_value = df['Employment'].mode()[0]
df['Employment'] = df['Employment'].fillna(most_frequent_value)

# Verify
missing_after_imputation = df['Employment'].isnull().sum()
print(f"Missing values in 'Employment' after imputation: {missing_after_imputation}")
```
